```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("retail_sales_superstore.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
region_summary = (
    df.groupby("Region")
      .agg(
          Total_Sales=("Sales", "sum"),
          Total_Profit=("Profit", "sum"),
          Order_Count=("Order_ID", "nunique")
      )
      .reset_index()
)

region_summary
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(region_summary["Region"], region_summary["Total_Sales"])

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(region_summary["Region"], region_summary["Total_Profit"])

plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")

plt.show()
```

```python
region_summary["Profit_Margin"] = (
    region_summary["Total_Profit"] / region_summary["Total_Sales"]
)

region_summary
```

```python
plt.figure(figsize=(8,5))

plt.bar(region_summary["Region"], region_summary["Profit_Margin"])

plt.title("Profit Margin by Region")
plt.xlabel("Region")
plt.ylabel("Profit Margin")

plt.show()
```

#  Project 3 Summary: Regional Performance Analysis

This project analyzed sales and profitability across regions.  
It showed that while sales may be similar, profit margins vary significantly, indicating that some regions are more cost-efficient and profitable than others.
