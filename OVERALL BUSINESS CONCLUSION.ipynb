# Retail Sales Business Analysis –Summary Report

This project performed end-to-end retail business analysis using sales transaction data to understand key drivers of profitability.

The analysis was divided into three parts:

1. Discount Impact Analysis  
2. Product Performance Analysis  
3. Regional Profitability Analysis  

# ✅ Overall Key Insights

### 1. Discount Impact
Discounting directly affects profitability. Higher discounts often reduce profit margins and can lead to losses.

### 2. Product Performance
Profit contribution varies across product categories and sub-categories. Certain product groups generate the majority of overall profit.

### 3. Regional Performance
Sales may appear balanced across regions, but profit margins differ significantly, showing that some regions operate more efficiently than others.

# Conclusion

Across discount, product, and regional analysis, profitability was found to depend more on pricing strategy, product mix, and regional efficiency than on sales volume alone.

## Business Recommendations
- Optimize discounting to protect profit margins  
- Focus on high-performing product categories  
- Improve strategies in low-profit regions to maximize overall business performance  

### Summary Statement
This project demonstrates how retail profitability can be improved through data-driven decisions across pricing, product strategy, and regional performance.
